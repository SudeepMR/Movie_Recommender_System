###Popularity-Based Movie Recommender System

```python
#import necessary libraries

import numpy as np
import pandas as pd
import difflib
```

```python
# loading the datasets from the csv file to dataframe

movies=pd.read_csv('/content/movies.csv')
ratings=pd.read_csv('/content/ratings.csv')
```

```python
# printing the first 5 rows of the movie dataframe

movies.head()
```

```python
# printing the first 5 rows of the ratings dataframe

ratings.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: x.replace("|", " "))
```

```python
# merging the datasets

df=pd.merge(ratings,movies, how='left',on='movieId')
df.head()
```

```python
# grouping title and genres with rating

ratings = pd.DataFrame(df.groupby(['title','genres'])['rating'].mean())
ratings.head()
```

```python
# creating a new column 'num of ratings' which holds the number of ratings given for the movie

ratings['num of ratings'] = pd.DataFrame(df.groupby(['title','genres'])['rating'].count())
ratings.head()
```

```python
# creating a pivot table

moviemat = df.pivot_table(index='userId',columns=['genres','title'],values='rating')
moviemat
```

```python
# grouping movies with most number of ratings

ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
# getting the movie genres name from the user

genres_name = input(' Enter your favourite movie genres name : ')
```

```python
# creating a list with all the movie names given in the dataset

list_of_all_genres = df['genres'].tolist()
list_of_all_genres = list(set(list_of_all_genres))
print(list_of_all_genres)
```

```python
# finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(genres_name, list_of_all_genres)
print(find_close_match)
```

```python
# selecting the first value

close_match = find_close_match[0]
print(close_match)
```

```python
moviemat = pd.DataFrame(moviemat)
moviemat
```

```python
# grouping the dataframe as per the input genre

genres_user_ratings = moviemat[close_match]
genres_user_ratings.head()
```

```python
# finding the correlation between the movies

similar_to_genres = moviemat.corrwith(genres_user_ratings)
```

```python
# creating a dataframe of similar movies and its correlation

corr_genres = pd.DataFrame(similar_to_genres,columns=['Correlation'])
corr_genres.dropna(inplace=True)
corr_genres.head()
```

```python
# sorting the dataframe in descending order according to correlation

corr_genres.sort_values('Correlation',ascending=False).head(10)
```

```python
# adding number of ratings and ratings columns to the dataframe

corr_genres = corr_genres.join(ratings['num of ratings'])
corr_genres = corr_genres.join(ratings['rating'])
corr_genres.head()
```

```python
# setting the threshold on number of ratings

rating_threshld = int(input(' Enter the Minimum reviews threshold : '))
```

```python
# sorting the dataframe in descending order according to ratings and correlation

final_df=corr_genres[corr_genres['num of ratings']>rating_threshld].sort_values(['rating','Correlation'],ascending=False).head()
final_df.head()
```

```python
# droping correlation column

final_df.drop('Correlation', axis=1, inplace=True)
```

```python
# taking number of recommendations from user

num_rec = int(input(' Enter Number of Recommendation : '))
```

```python
# final result

print('Movies suggested for you : \n')


print(final_df.head(num_rec+1))
```
